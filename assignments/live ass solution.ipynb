```python
# 3

np.array:

Purpose: Creates a new array, copying the data by default.
Behavior: Always creates a new ndarray object, unless the optional argument copy=False is used and the input is already an ndarray.
Use Case: When you need a new array and are okay with copying the data.
    
    np.asarray:

Purpose: Converts input to an array, but does not copy the data if the input is already an ndarray.
Behavior: If the input is already an ndarray, it returns the input unchanged. Otherwise, it creates a new ndarray.
Use Case: When you want to ensure the input is an array, but avoid unnecessary copying.
    
    np.asanyarray:

Purpose: Converts input to an array, but retains any subclass of ndarray.
Behavior: If the input is an instance of an ndarray subclass (e.g., matrix), it returns the input unchanged.
Use Case: When you want to convert input to an array but preserve any ndarray subclasses.
    
    deep copy:
        new object is created from old and any changes to new doesn't affect on old
        
    shallow copy:
        new objected created from old and any changes to new effect old
```

```python
# 14

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a time-series dataset
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.randint(-10, 30, size=100)
humidity = np.random.randint(30, 90, size=100)

df_timeseries = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Plot Temperature and Humidity with different y-axes
fig, ax1 = plt.subplots()

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='tab:red')
ax1.plot(df_timeseries['Date'], df_timeseries['Temperature'], color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel('Humidity', color='tab:blue')
ax2.plot(df_timeseries['Date'], df_timeseries['Humidity'], color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
# 15

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# a) Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# b) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# c) Overlay a line plot representing the normal distribution's probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# d) Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# e) Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
# 16

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Create a NumPy array with samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Plot PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Label axes and set title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()
```

```python
# 17

import seaborn as sns
import numpy as np
import pandas as pd

# Create two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Define a function to categorize points based on their quadrant
def categorize_point(row):
    if row['x'] > 0 and row['y'] > 0:
        return 'Q1'
    elif row['x'] < 0 and row['y'] > 0:
        return 'Q2'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Q3'
    else:
        return 'Q4'

df['quadrant'] = df.apply(categorize_point, axis=1)

# Create a scatter plot with Seaborn
sns.scatterplot(x='x', y='y', hue='quadrant', data=df)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.show()
```

```python
#19

from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
data = {'categories': categories, 'values': values}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values, color=['#%02x%02x%02x' % (int(val), 100, 100) for val in values]))

# Create a bar chart
p = figure(x_range=categories, title="Random Categorical Bar Chart", toolbar_location=None, tools="")

# Add bars
p.vbar(x='categories', top='values', width=0.9, source=source, line_color='white', fill_color='color')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label axes
p.xaxis.axis_label = "Category"
p.yaxis.axis_label = "Value"

# Show plot
show(p, notebook_handle=True)
```
