````python
def mean(numbers):
    return sum(numbers) / len(numbers) if numbers else 0
````

````python
def median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:
        return sorted_numbers[mid]
````

````python
from collections import Counter

def mode(numbers):
    frequency = Counter(numbers)
    max_freq = max(frequency.values())
    modes = [k for k, v in frequency.items() if v == max_freq]
    return modes
````

````python
def weighted_mean(values, weights):
    if len(values) != len(weights) or not values:
        return 0
    return sum(v * w for v, w in zip(values, weights)) / sum(weights)
````

````python
import math

def geometric_mean(numbers):
    if not numbers:
        return 0
    product = math.prod(numbers)
    return product ** (1 / len(numbers))
````

````python
def harmonic_mean(numbers):
    if not numbers or any(x == 0 for x in numbers):
        return 0
    return len(numbers) / sum(1 / x for x in numbers)
````

````python
def midrange(numbers):
    return (min(numbers) + max(numbers)) / 2 if numbers else 0
````

````python
def trimmed_mean(numbers, percentage):
    if not numbers or percentage <= 0 or percentage >= 100:
        return 0
    sorted_numbers = sorted(numbers)
    trim_count = int(len(sorted_numbers) * (percentage / 100))
    trimmed_numbers = sorted_numbers[trim_count:-trim_count]
    return mean(trimmed_numbers)
````

````python
def calculate_range(numbers):
    return max(numbers) - min(numbers) if numbers else 0
````

````python
def variance(numbers):
    mean_val = mean(numbers)
    return sum((x - mean_val) ** 2 for x in numbers) / (len(numbers) - 1) if numbers else 0

def standard_deviation(numbers):
    return variance(numbers) ** 0.5
````

````python
def interquartile_range(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    q1 = sorted_numbers[n // 4]
    q3 = sorted_numbers[3 * n // 4]
    return q3 - q1
````

````python
def coefficient_of_variation(numbers):
    std_dev = standard_deviation(numbers)
    mean_val = mean(numbers)
    return (std_dev / mean_val) * 100 if mean_val != 0 else 0
````

````python
def mean_absolute_deviation(numbers):
    mean_val = mean(numbers)
    return sum(abs(x - mean_val) for x in numbers) / len(numbers) if numbers else 0
````

````python
def quartile_deviation(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    q1 = sorted_numbers[n // 4]
    q3 = sorted_numbers[3 * n // 4]
    return (q3 - q1) / 2
````

````python
def range_based_coefficient_of_dispersion(numbers):
    range_val = calculate_range(numbers)
    mean_val = mean(numbers)
    return (range_val / mean_val) * 100 if mean_val != 0 else 0
````

````python
# 3

class DiscreteRandomVariable:
    def __init__(self, probabilities, values):
        self.probabilities = probabilities
        self.values = values
        if len(probabilities) != len(values):
            raise ValueError("Length of probabilities and values must be the same.")
        if not all(0 <= p <= 1 for p in probabilities) or not abs(sum(probabilities) - 1) < 1e-9:
            raise ValueError("Probabilities must sum to 1 and be between 0 and 1.")
    
    def expected_value(self):
        return sum(p * v for p, v in zip(self.probabilities, self.values))
    
    def variance(self):
        mean = self.expected_value()
        return sum(p * (v - mean) ** 2 for p, v in zip(self.probabilities, self.values))

# Example usage:
probabilities = [0.1, 0.2, 0.3, 0.4]
values = [1, 2, 3, 4]
rv = DiscreteRandomVariable(probabilities, values)
print("Expected Value:", rv.expected_value())
print("Variance:", rv.variance())
````

````python
#5 
import numpy as np

def generate_samples(distribution, *params, size=1000):
    if distribution == 'binomial':
        return np.random.binomial(params[0], params[1], size)
    elif distribution == 'poisson':
        return np.random.poisson(params[0], size)
    else:
        raise ValueError("Unsupported distribution")

def mean_and_variance(samples):
    mean_val = np.mean(samples)
    variance_val = np.var(samples)
    return mean_val, variance_val

# Example usage for binomial distribution
binomial_samples = generate_samples('binomial', 10, 0.5)
print("Binomial Mean and Variance:", mean_and_variance(binomial_samples))

# Example usage for Poisson distribution
poisson_samples = generate_samples('poisson', 5)
print("Poisson Mean and Variance:", mean_and_variance(poisson_samples))
````

````python
#8
import numpy as np

def normal_pdf(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Example usage:
mu = 0  # Mean
sigma = 1  # Standard deviation
x = 1.0  # Value at which to compute the PDF
pdf_value = normal_pdf(x, mu, sigma)
print(f"PDF value at x={x}: {pdf_value}")
````

````python
#9
def exponential_cdf(x, lambda_):
    return 1 - np.exp(-lambda_ * x)

# Example usage:
lambda_ = 1  # Rate parameter (1/mean)
x = 2.0  # Value at which to compute the CDF
cdf_value = exponential_cdf(x, lambda_)
print(f"CDF value at x={x}: {cdf_value}")
````

;M A company wants to test iN a new website layout leads to a higher conversion rate (percentage oN visitors

To ge<erate the data use the followi<g comma<d:

```python

import <umpy as <p

# 50 purchases out of 1000 visitors

old_layout = <p.array([1] * 50 + [0] * 950)

# 70 purchases out of 1000 visitors  

<ew_layout = <p.array([1] * 70 + [0] * 930)

  ```

Apply z-test to fi<d which layout is successful.



````python
import numpy as np
from scipy import stats

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate conversion rates
p_old = np.mean(old_layout)
p_new = np.mean(new_layout)

# Calculate sample sizes
n_old = len(old_layout)
n_new = len(new_layout)

# Calculate the pooled proportion
p_pooled = (np.sum(old_layout) + np.sum(new_layout)) / (n_old + n_new)

# Calculate the standard error
standard_error = np.sqrt(p_pooled * (1 - p_pooled) * (1 / n_old + 1 / n_new))

# Calculate the z-statistic
z_statistic = (p_new - p_old) / standard_error

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Print the results
print(f"Old Layout Conversion Rate: {p_old:.4f}")
print(f"New Layout Conversion Rate: {p_new:.4f}")
print(f"Z-Statistic: {z_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the old and new layouts.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the old and new layouts.")
````

```python

after_program = <p.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

```

Use z-test to fi<d if the claims made by tutor are true or false.



````python
import numpy as np
from scipy import stats

# Generate the data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate the differences
differences = after_program - before_program

# Calculate the mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)
n = len(differences)

# Calculate the t-statistic
t_statistic = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=n-1))

# Print the results
print(f"Mean Difference: {mean_diff:.4f}")
print(f"Standard Deviation of Differences: {std_diff:.4f}")
print(f"T-Statistic: {t_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The tutoring program significantly improves exam scores.")
else:
    print("Fail to reject the null hypothesis: There is no significant improvement in exam scores from the tutoring program.")
````

````python
#15
import numpy as np
from scipy import stats

def perform_ab_test(layout_a_clicks, layout_b_clicks):
    # Convert lists to numpy arrays
    a = np.array(layout_a_clicks)
    b = np.array(layout_b_clicks)
    
    # Calculate means and standard deviations
    mean_a = np.mean(a)
    mean_b = np.mean(b)
    std_a = np.std(a, ddof=1)
    std_b = np.std(b, ddof=1)
    n_a = len(a)
    n_b = len(b)
    
    # Calculate the t-statistic
    pooled_var = (((n_a - 1) * std_a**2) + ((n_b - 1) * std_b**2)) / (n_a + n_b - 2)
    se = np.sqrt(pooled_var * (1/n_a + 1/n_b))
    t_statistic = (mean_a - mean_b) / se
    
    # Calculate degrees of freedom
    df = n_a + n_b - 2
    
    # Calculate the p-value
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))
    
    # Print results
    print(f"Mean of Layout A: {mean_a:.2f}")
    print(f"Mean of Layout B: {mean_b:.2f}")
    print(f"Standard Deviation of Layout A: {std_a:.2f}")
    print(f"Standard Deviation of Layout B: {std_b:.2f}")
    print(f"T-Statistic: {t_statistic:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-Value: {p_value:.4f}")

    # Interpretation
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference in click-through rates between the two layouts.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference in click-through rates between the two layouts.")

# Example usage
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

perform_ab_test(layout_a_clicks, layout_b_clicks)
````

````python
#16
import numpy as np
from scipy import stats

def analyze_drug_effect(existing_drug_levels, new_drug_levels):
    # Convert lists to numpy arrays
    existing = np.array(existing_drug_levels)
    new = np.array(new_drug_levels)
    
    # Calculate means and standard deviations
    mean_existing = np.mean(existing)
    mean_new = np.mean(new)
    std_existing = np.std(existing, ddof=1)
    std_new = np.std(new, ddof=1)
    n_existing = len(existing)
    n_new = len(new)
    
    # Calculate the t-statistic
    pooled_var = (((n_existing - 1) * std_existing**2) + ((n_new - 1) * std_new**2)) / (n_existing + n_new - 2)
    se = np.sqrt(pooled_var * (1/n_existing + 1/n_new))
    t_statistic = (mean_existing - mean_new) / se
    
    # Calculate degrees of freedom
    df = n_existing + n_new - 2
    
    # Calculate the p-value
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))
    
    # Print results
    print(f"Mean Cholesterol Level (Existing Drug): {mean_existing:.2f}")
    print(f"Mean Cholesterol Level (New Drug): {mean_new:.2f}")
    print(f"Standard Deviation (Existing Drug): {std_existing:.2f}")
    print(f"Standard Deviation (New Drug): {std_new:.2f}")
    print(f"T-Statistic: {t_statistic:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-Value: {p_value:.4f}")

    # Interpretation
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: The new drug significantly reduces cholesterol levels compared to the existing drug.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference in cholesterol levels between the new drug and the existing drug.")

# Example usage
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

analyze_drug_effect(existing_drug_levels, new_drug_levels)
````

````python
import numpy as np
from scipy import stats

def analyze_intervention_effect(pre_intervention_scores, post_intervention_scores):
    # Convert lists to numpy arrays
    pre_scores = np.array(pre_intervention_scores)
    post_scores = np.array(post_intervention_scores)
    
    # Calculate the differences between post and pre-intervention scores
    differences = post_scores - pre_scores
    
    # Calculate mean and standard deviation of differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)
    n = len(differences)
    
    # Calculate the t-statistic
    t_statistic = mean_diff / (std_diff / np.sqrt(n))
    
    # Calculate degrees of freedom
    df = n - 1
    
    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))
    
    # Print results
    print(f"Mean Difference: {mean_diff:.2f}")
    print(f"Standard Deviation of Differences: {std_diff:.2f}")
    print(f"T-Statistic: {t_statistic:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-Value: {p_value:.4f}")

    # Interpretation
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: The educational intervention significantly improved math scores.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference in math scores before and after the intervention.")

# Example usage
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

analyze_intervention_effect(pre_intervention_scores, post_intervention_scores)
````

````python
#21
import numpy as np
import pandas as pd
from scipy import stats

# Generate synthetic data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Print results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between age groups and voter preferences.")
````
